```python
from sklearn.tree import export_graphviz, DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

from IPython.display import Image

import pydotplus
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston.feature_names
```

```python
hasattr(boston, 'target_names'), hasattr(boston, 'feature_names')
```

```python
cancer = load_breast_cancer()
```

```python
cancer.target_names, cancer.feature_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=0)
```

```python
tree = DecisionTreeClassifier(max_depth=8)
tree.fit(X_train, y_train)
```

```python
cancer.feature_names
```

```python
tree_dot = export_graphviz(tree, out_file=None, feature_names=cancer.feature_names,)
```

```python
graph = pydotplus.graph_from_dot_data(tree_dot)  
print(type(graph.create_png()))
Image(graph.create_png())
```
