```python
from IPython.display import display, HTML
from graphviz import Digraph
from urllib import parse
import base64
```

```python
dot = Digraph()
# Add nodes
dot.node('1', label='a')
dot.node('3')
dot.node('2')
dot.node('5')

# Add edges
dot.edges(['12', '13', '35'])

# Visualize the graph
print(type(dot))
dot
```

# Show svg xml

```python
data = dot._repr_svg_()
data = ''.join(data.split('\n')[6:]) # strip
data
```

------------------

# 方式一: 直接使用svg标签

```python
# <svg width="134pt" height="188pt" ... >
#    .....
# </svg>
HTML(data=data)
```

## 方法三: 使用img标签, type: data:image/svg+xml;base64
